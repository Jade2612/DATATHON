```python
#câu 2
import pandas as pd
import numpy as np 
df = pd.read_csv('products.csv')
df.info() 
copy_df = df.copy() 
copy_df['profit'] = (copy_df['price'] - copy_df['cogs'])/copy_df['price']
copy_df = copy_df.groupby('segment')['profit'].mean().sort_values(ascending=False)
copy_df.head(20)
```

```python
#câu 3
df1 = pd.read_csv('returns.csv')
df2 = pd.read_csv('products.csv')
df2 = df2[df2['category'] == 'Streetwear']
copy_df = df1.merge(df2, on='product_id', how='inner')
copy_df.info()
copy_df.head(20)
copy_df = copy_df.groupby('return_reason').size().sort_values(ascending=False)
copy_df.head(20)
```

```python
#câu 4
df = pd.read_csv('web_traffic.csv')
df.info() 
df.head(10)
df = df.groupby('traffic_source')['bounce_rate'].mean().sort_values(ascending=True)
df.head(20)
```

```python
#câu 5
df = pd.read_csv('order_items.csv')
df.info() 
df.head(10)
result = (df['promo_id'].notna().sum())/df.shape[0]*100
print(result)
```

```python
#Câu 6
df = pd.read_csv('customers.csv')
df.info()  
df.head(10)
df3 = pd.read_csv('orders.csv')
df3.info()
merged = df.merge(df3, on='customer_id', how='inner')
merged.head(20)
result = merged.groupby('age_group')['order_id'].size() / merged.groupby('age_group')['customer_id'].nunique()
result.head(20)
```

```python
#câu 8
df = pd.read_csv('orders.csv')
df = df[df['order_status'] == 'cancelled']
df = df.groupby('payment_method').size().sort_values(ascending=False)
print(df)
```

```python
#câu 9 
return_df = pd.read_csv('returns.csv')
product = pd.read_csv('products.csv')
order = pd.read_csv('order_items.csv')
return_df.info()
return_df = pd.merge(return_df, product, on='product_id', how='inner')
order = pd.merge(order, product, on='product_id', how='inner')
num = return_df.groupby('size').size()
deno = order.groupby('size').size()
result = num/deno
print(result)
```
